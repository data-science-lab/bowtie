```python
import networkx as nx
from numpy import *
```

```python
with open("D:/Data/ndwww/www.dat", "rb") as f:
    www = f.readlines()
```

```python
G = nx.DiGraph()
for i in www:
    start =  i.split(" ")[0]
    end = i.split(" ")[1].split("\n")[0]
    G.add_edge(start, end)
```

```python
print nx.info(G)
```

```python
K = len(G.edges())
N = len(G.nodes())
averageki = 2*K / float(N)
print averageki
```

```python
# 将G转化成无向图
nG = G.to_undirected()
print nx.info(nG)
```

```python
# 计算K和 (N*(N - 1)/2)的大小比较，是否是<<
float(K) / (N*(N - 1)/2)
```

```python
D = nx.average_shortest_path_length(nG)
```

```python
C = nx.average_clustering(nG)
```
